## Average percentage of adults who engage in no leisure-time physical activity across different income levels

```python

sorted_data = df.sort_values(by='Income')


physical_activity_data = sorted_data[sorted_data['Question'] == "Percent of adults who engage in no leisure-time physical activity"]


physical_activity_data = physical_activity_data.dropna(subset=['Data_Value'])

mean_percentage_by_income = physical_activity_data.groupby('Income')['Data_Value'].mean()
formatted_percentages_income = mean_percentage_by_income.round(2).map("{:.2f}%".format)
formatted_percentages_income.round(2)
```

## Percent of adults aged 18 years and older who have an overweight classification

```python
overweight_data = df[df['Question'] == "Percent of adults aged 18 years and older who have an overweight classification"]
overweight_data = overweight_data.dropna(subset=['Data_Value'])
percentage_overweight_by_ethnicity = overweight_data.groupby('Race/Ethnicity')['Data_Value'].mean()
print(overweight_data['Question'].iloc[0])

percentage_overweight_by_ethnicity.round(2)
formatted_percentages = percentage_overweight_by_ethnicity.round(2).map("{:.2f}%".format)
formatted_percentages
```

## Percent of adults aged 18 years and older who have obesity

```python
obese_data = df[df['Question'] == "Percent of adults aged 18 years and older who have obesity"]

obese_data_filtered = obese_data.dropna(subset=['Data_Value'])
percentage_obese_by_ethnicity = obese_data_filtered.groupby('Race/Ethnicity')['Data_Value'].mean()

print(obese_data_filtered['Question'].iloc[0])
formatted_obese_percentages = percentage_obese_by_ethnicity.round(2).map("{:.2f}%".format)
formatted_obese_percentages
```

## Average percentage of people achieve at least 150 minutes a week of moderate-intensity aerobic physical activity or 75 minutes a week of vigorous-intensity aerobic physical activity and engage in muscle-strengthening activities on 2 or more days a week

```python
physical_activity_data = df[df['Question'] == "Percent of adults who achieve at least 150 minutes a week of moderate-intensity aerobic physical activity or 75 minutes a week of vigorous-intensity aerobic physical activity and engage in muscle-strengthening activities on 2 or more days a week"]
physical_activity_data = physical_activity_data.dropna(subset=['Data_Value'])

mean_percentage_by_location = physical_activity_data.groupby('LocationDesc')['Data_Value'].mean()

mean_percentage_by_location_sorted = mean_percentage_by_location.sort_values(ascending=False)

mean_percentage_by_location_sorted_formatted= mean_percentage_by_location_sorted.round(2).map("{:.2f}%".format)
print("Percent of adults who achieve at least 150 minutes a week of moderate-intensity aerobic physical activity or 75 minutes a week of vigorous-intensity aerobic physical activity and engage in muscle-strengthening activities on 2 or more days a week")
mean_percentage_by_location_sorted_formatted.round(2)
```

#### Bar Chart - Physical Activity Trends Over Years
This bar chart visualizes the trend of physical activity levels over the years. It helps to understand how public engagement in physical activity has evolved.

```python
import matplotlib.pyplot as plt
import pandas as pd

yearly_activity = df.groupby('YearStart')['Data_Value'].mean().reset_index()

# Creating the bar chart
plt.figure(figsize=(10, 6))
plt.bar(yearly_activity['YearStart'], yearly_activity['Data_Value'], color='skyblue')
plt.title('Average Physical Activity Over Years')
plt.xlabel('Year')
plt.ylabel('Average Physical Activity (%)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Pie Chart - Physical Activity Distribution by Race/Ethnicity
This pie chart breaks down the proportion of physical activity data entries by different races and ethnicities, providing insight into data representation.


```python
import matplotlib.pyplot as plt

if not sorted_percentage_active_by_race.empty:
    # Plotting the pie chart
    plt.figure(figsize=(10, 8))
    plt.pie(sorted_percentage_active_by_race, labels=sorted_percentage_active_by_race.index, autopct='%1.1f%%', startangle=90)
    plt.title('Distribution of Physical Activity by Race/Ethnicity')
    plt.show()
else:
    print("No data available for plotting. Check the filtering and data processing steps.")
```

#### Scatter Plot - Relationship Between Income and Physical Activity
This scatter plot examines the potential correlation between income levels and physical activity percentages, exploring socioeconomic factors in health behavior.


```python
import matplotlib.pyplot as plt
import pandas as pd
df['Income'] = df['Income'].astype(str)

plt.figure(figsize=(12, 8))
plt.scatter(df['Income'], df['Data_Value'], alpha=0.6, c='green')
plt.title('Physical Activity vs. Income Levels')
plt.xlabel('Income Level')
plt.ylabel('Physical Activity (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#### Stacked Area Chart - Physical Activity Trends by Gender Over Years
This stacked area chart visualizes how physical activity levels have differed between genders over the years, providing a gendered perspective on public health engagement.


```python
gender_activity = df.pivot_table(index='YearStart', columns='Gender', values='Data_Value', aggfunc='mean')

plt.figure(figsize=(12, 8))
plt.stackplot(gender_activity.index, gender_activity['Male'], gender_activity['Female'], labels=['Male', 'Female'], colors=['lightblue', 'pink'])
plt.title('Physical Activity Trends by Gender Over Years')
plt.xlabel('Year')
plt.ylabel('Average Physical Activity (%)')
plt.legend(loc='upper left')
plt.show()
```

#### Line Chart - Trends in Obesity Rates Over Years by Location
This line chart displays the trend of obesity rates over the years in different locations. It highlights how obesity trends have evolved, allowing for comparisons across regions

```python
import matplotlib.pyplot as plt
import pandas as pd

obesity_data = df[df['Topic'] == 'Obesity / Weight Status']
obesity_data = obesity_data[obesity_data['Question'] == 'Percent of adults aged 18 years and older who have obesity']
obesity_trends = obesity_data.groupby(['YearStart', 'LocationDesc'])['Data_Value'].mean().unstack()


plt.figure(figsize=(14, 8))
for location in obesity_trends.columns:
    plt.plot(obesity_trends.index, obesity_trends[location], marker='o', label=location)

plt.title('Trends in Obesity Rates Over the Years by Location')
plt.xlabel('Year')
plt.ylabel('Average Obesity Rate (%)')
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
